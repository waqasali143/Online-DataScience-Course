```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("netflix_data.csv")
data.head()
```

```python
data["cast"].replace(np.nan, "None",inplace=True)
data.head()
```

```python
def cast_len(val):
    if val == "None":
        return 0
    else:
        return len(str(val).split(", "))
data["Num_of_cast"]=data["cast"].apply(cast_len)
data.head()
```

```python
data.groupby(["type","genre"]).count()
```
